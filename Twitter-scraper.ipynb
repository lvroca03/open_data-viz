# Twitter scrapper

```python
import asyncio
from twscrape import API, gather
import pandas as pd 
```

```python
id=[]
username=[]
tweets=[]
location=[]
displayname=[]
date=[]
```

```python
async def main():
    api = API()
    await api.pool.delete_accounts(["lvroca1"])
    #colocar sus credenciales
    await api.pool.add_account("usuario twitter", "contraseña twitter","correo","contraseña correo")

    await api.pool.login_all()

    tweet_counter = 0 

    await gather(api.search("guerra israel gaza", limit= 500, kv={"product": "Top"}))

    async for tweet in api.search("guerra israel gaza"):
        print(tweet.id,tweet.rawContent )
        id.append(tweet.id)
        location.append(tweet.user.location)
        displayname.append(tweet.user.displayname)
        date.append(tweet.date)
        username.append(tweet.user.username) 
        tweets.append(tweet.rawContent) 
        
        tweet_counter += 1

        if tweet_counter >= 2000:
            print("se escrapaeron exitosamente 2000 tweets.")
            break
    

if __name__ == "__main__":
    if asyncio.get_event_loop().is_running():
        asyncio.create_task(main())
    else:
        asyncio.run(main())
```

```python
tweets_guerra=pd.DataFrame({"id":id,"usuario":username,"tweet":tweets,"ubicacion":location,"nombre_de_pila":displayname,"fecha":date})
tweets_guerra["fecha"] = tweets_guerra["fecha"].astype(str)
tweets_guerra.to_excel('tweets_guerra.xlsx', index=False)
```
